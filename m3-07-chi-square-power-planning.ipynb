```python
np.random.seed(42)

observed = np.array([120, 85, 110, 85])

expected = np.array([100, 100, 100, 100]) #because 25% of 400 is 100

chi2_stat, p_value = st.chisquare(observed, f_exp=expected)

df = len(observed) - 1

print(f"Chi-square statistic: {chi2_stat:.2f}")
print(f"Degrees of freedom: {df}")
print(f"P-value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("Reject H0: Class preferences are not equal.")
else:
    print("Fail to reject H0: No evidence against equal preferences.")

"""
Warrior (+20) and Rogue (+10) are over-represented.
Mage (-15) and Healer (-15) are under-represented compared to expected counts.
"""
```

```python
np.random.seed(42)

df = {
    "Free" : [90, 110],
    "Basic" : [60, 140],
    "Premium" : [30, 170]
}

row_names = ("Churned", "Retained")

df = pd.DataFrame(df, index = row_names)

print(df)

print("")

chi2_stat, p_value, dof, expected = st.chi2_contingency(df)

print(f"Chi-square statistic: {chi2_stat:.2f}")
print(f"P-value: {p_value:.20f}")
print(f"Degrees of freedom: {dof}")
print("Expected frequencies:")
print(expected)

"""
There is a strong evidence of an association between subscription tier and churn, because p < 0,05.
"""
```

```python
def cramers_v(chi2, n, min_dim):
    """
    Returns Cramér's V.
    
    Parameters
    ----------
    chi2 : float
        Chi-square statistic.
    n : int
        Total number of observations.
    min_dim : int
        min(rows, cols) - 1 of the contingency table.
    """
    return (chi2 / (n * min_dim))**0.5

n = df.values.sum()

min_dim = min(df.shape) - 1  # min(rows, cols) - 1

v = cramers_v(chi2_stat, n, min_dim)
print(f"Cramér's V: {v:.3f}")

if v < 0.1:
    effect = "negligible"
elif v < 0.3:
    effect = "small"
elif v < 0.5:
    effect = "medium"
else:
    effect = "large"

print(f"Effect size: {effect}")
```
